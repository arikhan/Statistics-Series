```python
# Let's populate our sample set by using mean and standard deviation!
mu = 0 # Here is my mean parameter
sigma = 1 # Here is my standard deviation parameter.
sample_size = 1000 # Parameter for how many samples that we will get
normal_samples = np.random.normal(mu, sigma, sample_size)
normal_samples
```

```python
# Now let's see our lovely normal distribution!
sns.histplot(data=normal_samples, kde=True)
```

```python
# Let's populate our sample set by using low and high values!
low = 0 # Here is my low parameter
high = 1 # Here is my high parameter.
sample_size = 1000 # Parameter for how many samples that we will get
uniform_samples = np.random.uniform(mu, sigma, sample_size)
uniform_samples
```

```python
# Now let's see our lovely uniform distribution!
g = sns.distplot(uniform_samples)
```
